```python
file_path = "/content/P1_data.txt"
def exercide3(file_path):
    word_count = {}

    with open(file_path, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                # word = word.strip().lower()
                word = word.strip('.,!?:;"()[]').lower()

                if word in word_count:
                    word_count[word] += 1
                else:
                    word_count[word] = 1

    return word_count

exercide3(file_path)
```
