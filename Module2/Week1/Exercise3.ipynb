```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/advertising.csv')
A = df['Sales'].mean()
scores = np.where(df['Sales'] > A, 'Good', np.where(df['Sales'] < A, 'Bad', 'Average'))
print(scores[7:10])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/advertising.csv')
A = df.iloc[(df['Sales'] - df['Sales'].mean()).abs().argsort()[:1]]['Sales'].values[0]
scores = np.where(df['Sales'] > A, 'Good', np.where(df['Sales'] < A, 'Bad', 'Average'))
print(scores[7:10])
```
